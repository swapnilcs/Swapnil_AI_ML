```python
# importing basic libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
df = pd.read_excel('flight_price.xlsx')
```

```python
df.head()
```

```python
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
## Convert Date, Month and Year from object type to int
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
df.info()
```

```python
## Drop col. Date_of_Journey
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Arrival_Time']
```

```python
df['Arrival_Time'].str.split(':').str[0]
```

```python
df['Arrival_Time'].str.split(':').str[1].str.split(' ').str[0]
```

```python
df['Arrival_Hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_Min'] = df['Arrival_Time'].str.split(':').str[1].str.split(' ').str[0]
```

```python
df.head()
```

```python
## Drop col. Arrival_Time
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
## Crovert Arrival_Hour and Arrival_Min to int
df['Arrival_Hour'] = df['Arrival_Hour'].astype(int)
df['Arrival_Min'] = df['Arrival_Min'].astype(int)
```

```python
df.info()
```

```python
df.head()
```

```python
df['Departure_Hour'] = df['Dep_Time'].str.split(':').str[0]
df['Departure_Min'] =  df['Dep_Time'].str.split(':').str[1]
```

```python
## Drop Dep_Time col
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
## Convert Departure_Hour and Departure_Min to int
df['Departure_Hour'] = df['Departure_Hour'].astype(int)
df['Departure_Min'] = df['Departure_Min'].astype(int)
```
