```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read file from the location
df = pd.read_csv('height-weight.csv')
df.head()
```

```python
## Scatter plot to check the relation between out independent and dependent values
plt.scatter(df['Weight'],df['Height'])
plt.xlabel("Weight")
plt.ylabel("Height")
plt.show()
```

```python
# find Correlation
df.corr()
```

```python
# seaborn for visualization
import seaborn as sns
sns.pairplot(df)
plt.show()
```

```python
## Divide our data into independent and dependent features
X = df[['Weight']]  # Should be in df or 2D array
y = df['Height']   
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train.shape
```

```python
## standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_train
```

```python
X_test = scaler.transform(X_test)
X_test
```

```python
# Apply Simple Linear Regression
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
regression.fit(X_train,y_train)
```

```python
print('Slope or Coefficient is :- ',regression.coef_)
print('Inteecept is :-', regression.intercept_)
```

```python
# plot best fit line w.r.t. training data
plt.scatter(X_train,y_train)
plt.plot(X_train,regression.predict(X_train))
plt.show()
```

```python
# Prediction for test data
y_pred = regression.predict(X_test)
```

```python
# Performance metrics
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
mse = mean_squared_error(y_test,y_pred)
mae = mean_absolute_error(y_test,y_pred)
rmse = np.sqrt(mse)
print(mse)
print(mae)
print(rmse)
```

```python
# R squared 
from sklearn.metrics import r2_score
```

```python
score = r2_score(y_test,y_pred)
score
```
